fock loss distribution

```python
npzfile = np.load('loss.npz')
loss = npzfile['data']
```

```python
 # 计算数据的对数  
log_data = np.log10(abs(loss))  
  
# 计算直方图的bins和频数  
hist, bins = np.histogram(log_data, bins=20)  
  
# 计算bins的中点，以便在柱状图中准确表示  
bin_centers = 0.5 * (bins[1:] + bins[:-1])  
  
# 绘制柱状图  
plt.figure(figsize=(12,7))
plt.bar(bin_centers, hist, width=(bins[1] - bins[0]))  
  
# # 设置横坐标为对数刻度（如果你想显示原始数据的刻度）  
# plt.xticks(ticks=bin_centers, labels=np.round(10**bin_centers, 2), rotation=45)  
  
# 设置横坐标和纵坐标的标签  
plt.xlabel('Loss mae (log scale) 1e-x')  
plt.ylabel('Frequency')  
  
# 显示图形  
plt.show()  
```
